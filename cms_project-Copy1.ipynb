```python
class complaint:
    def __init__(self,complaint_id,title,description,status):
        self.complaint_id=complaint_id
        self.title=title
        self.description=description
        self.status=status
    def __str__(self):
        return f"complaint ID:{self.complaint_id}/nTitle:{self.title}/nDesription:{self.description}/nStatus:{self.status}/n"
class complaintmanagementsystem:
    def __init__(self):
        self.complaints=[]
    def register_complaint(self,complaint_id,title,description,status):
        self.complaints.append(complaint(complaint_id,title,description,status))
    def view_complaint(self):
        return self.complaints
    def update_complaint(self,complaint_id,new_status):
        for complaint in self.complaints:
            if complaint.complaint_id==complaint_id:
                complaint.status=new_status
                break
            else:
                print("complaint not found")
def main():
    cms=complaintmanagementsystem()
    while True:
        print(1,"Register complaint")
        print(2,"view complaints")
        print(3,"update complaints")
        print(4,"Exit")
        choice=int(input("Enter your choices(1/2/3/4):"))
        if choice==1:
            complaint_id=int(input("Enter complaint id:"))
            title=input("Enter title:")
            description=input("Enter description:")
            status=input("Enter status:")
            cms.register_complaint(complaint_id,title,description,status)
            print("complaint registerd successfully")
        elif choice==2:
            print("current complaints:")
            for complaint in cms.view_complaint():
                print(complaint)
        elif choice==3:
            complaint_id=int(input("Enter complaint id:"))
            new_status=input("Enter new status:")
            cms.update_complaint(complaint_id,new_status)
            print("complaint updated successfully")
        elif choice==4:
            print("Exit complaint management system")
            break
        else:
            print("Invalid choice.please try again")
if __name__ =="__main__":
    main()
    
                
            
```
